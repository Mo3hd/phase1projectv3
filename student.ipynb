```python
import pandas as pd
```

```python
file_path = '/Users/mac/Downloads/AviationData.csv'
```

```python
df = pd.read_csv(file_path,encoding='latin1')

df.head()
```

```python
# Check for missing values in the DataFrame
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Dropping columns with most missing values

# Specify the columns you want to drop in a list
columns_to_drop = ['Latitude', 'Longitude', 'Airport.Code', 'Airport.Name', 'Aircraft.Category', 'FAR.Description', 'Schedule', 'Air.carrier', 'Publication.Date']


# Drop the specified columns
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.head(25)
```

```python
# Filling in the missing values of the remaining columns with the mean
# Fill missing values with the mean of each column
df.fillna(df.mean(), inplace=True)
```

```python
df.head(10)
```

```python
# Check for duplicates in the entire DataFrame
duplicate_rows = df[df.duplicated()]

# If you want to check duplicates based on specific columns, you can specify them like this:
# duplicate_rows = df[df.duplicated(subset=['Column1', 'Column2'])]

# Print out the duplicate rows
print(duplicate_rows)
```

## Visualizations

```python
# Bar Chart for Total Fatal Injuries, Total Serious Injuries, Total Minor Injuries, and Total Uninjured

import matplotlib.pyplot as plt

# Data for the bar chart
labels = ['Total Fatal Injuries', 'Total Serious Injuries', 'Total Minor Injuries', 'Total Uninjured']
values = [df['Total.Fatal.Injuries'].sum(), df['Total.Serious.Injuries'].sum(), df['Total.Minor.Injuries'].sum(), df['Total.Uninjured'].sum()]

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(labels, values, color='skyblue')
plt.xlabel('Injury Type')
plt.ylabel('Total Count')
plt.title('Distribution of Injuries')
plt.show()
```

```python
# Pie Chart for Injury Severity

# Data for the pie chart
labels = df['Injury.Severity'].value_counts().index
sizes = df['Injury.Severity'].value_counts().values

# Create pie chart
plt.figure(figsize=(15, 15))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.axis('equal')
plt.title('Distribution of Injury Severity')
plt.show()
```

```python
# Create a histogram for Total Fatal Injuries
plt.figure(figsize=(12, 6))
plt.hist(df['Total.Fatal.Injuries'], bins=20, color='lightblue')
plt.xlabel('Total Fatal Injuries')
plt.ylabel('Frequency')
plt.title('Histogram of Total Fatal Injuries')
plt.show()
```

## 3. Relationships between different variables using scatter plots or correlation matrices

```python
# Create scatter plot for the specified columns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df['Total.Fatal.Injuries'], df['Total.Serious.Injuries'], color='green', alpha=0.7)
plt.xlabel('Total Fatal Injuries')
plt.ylabel('Total Serious Injuries')
plt.title('Scatter Plot between Total Fatal Injuries and Total Serious Injuries')
plt.show()
```

```python
# Create scatter plot for 'Total.Minor.Injuries' and 'Total.Uninjured'
plt.figure(figsize=(10, 6))
plt.scatter(df['Total.Minor.Injuries'], df['Total.Uninjured'], color='green', alpha=0.7)
plt.xlabel('Total Minor Injuries')
plt.ylabel('Total Uninjured')
plt.title('Scatter Plot between Total Minor Injuries and Total Uninjured')
plt.show()
```
